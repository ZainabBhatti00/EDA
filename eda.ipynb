```python
# cleaning & filtering the data
#  firstly find missing values in dataset
Ks.isnull()
```

```python
Ks.isnull().sum()
# age has 177 missing/NaN values & deck has 688 NaN so we will remove deck column first
```

```python
#removing missing value data or cleaning data
Kclean= pd.DataFrame(Ks.drop(["deck"], axis =1))
Kclean.head()
```

```python
Kclean.isnull().sum()
```

```python
Kclean.shape
Kclean.head()
```

```python
#dropna() method allows the user to analyze and drop Rows with Null values
#we use it here to drop missing values from AGE column

Kclean = Kclean.dropna()
```

```python
Kclean.describe()
```

```python
sns.boxplot(x= "sex", y="age", data= Kclean)
```

```python
sns.boxplot( y="age", data= Kclean)# outliar exists in age columns
```

```python
sns.displot(Kclean["age"])
```

```python
#out liars removal
Kclean["age"].mean()
```

```python
# as we can see from above boxplot out liars are above 70
Kclean= Kclean[Kclean['age'] < 68]
Kclean.head()

```

```python
Kclean.shape# we can see outliars have been removed
```

```python
sns.boxplot( y="age", data= Kclean)# outliar have been removed from boxplot
```
